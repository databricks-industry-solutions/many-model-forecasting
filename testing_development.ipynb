```python
from mmf_sa import run_forecast
import uuid
```

```python
from pyspark.sql import SparkSession

spark = SparkSession \
    .builder \
    .appName("MMF DEVELOPMENT") \
    .getOrCreate()
```

```python
run_id = str(uuid.uuid4())
user = "agreharshit610@gmail.com"
```

```python

active_models = ["NeuralForecastLSTM"]
table_name  = "yahoo_stock"
catalog = "workspace"
db = "mmf_development"
scoring_output = "yahoo_stock_scoring_output3"
evaluation_output = "yahoo_stock_evaluation_output3"


run_forecast(
    spark=spark,
    train_data=f"{catalog}.{db}.{table_name}",
    scoring_data=f"{catalog}.{db}.{table_name}",
    scoring_output=f"{catalog}.{db}.{scoring_output}",
    evaluation_output=f"{catalog}.{db}.{evaluation_output}",
    model_output=f"{catalog}.{db}",
    group_id="group_id",
    date_col="Date",
    target="Open",
    freq="D",
    prediction_length=7,
    backtest_months=2,
    stride=7,
    metric="smape",
    train_predict_ratio=1,
    data_quality_check=False,
    resample=True,
    active_models=active_models,
    experiment_path=f"/Users/{user}/mmf_development/yahoo_stock3",
    use_case_name="yahoo_stock3",
    run_id=run_id,
    accelerator="cpu",
)
```
