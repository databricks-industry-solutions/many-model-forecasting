# MMF GPU Models Orchestrator
Auto-generated by mmf-agent. Runs each model in a fresh Python process via `dbutils.notebook.run()`.

```python
catalog = "{catalog}"
schema = "{schema}"
use_case = "{use_case}"
train_table = "{train_table}"
freq = "{freq}"
prediction_length = {prediction_length}
backtest_length = {backtest_length}
stride = {stride}
metric = "{metric}"
group_id = "{group_id}"
date_col = "{date_col}"
target = "{target}"
num_nodes = {num_nodes}

active_models = {active_models}
```

```python
import uuid

run_id = str(uuid.uuid4())
user = spark.sql("SELECT current_user() AS user").collect()[0]["user"]

print(f"Orchestrator starting: {len(active_models)} models")
print(f"Run ID: {run_id}")
print(f"Models: {active_models}")

for i, model in enumerate(active_models, 1):
    print(f"\n[{i}/{len(active_models)}] Running model: {model}")
    dbutils.notebook.run(
        "./run_gpu",
        timeout_seconds=0,
        arguments={
            "catalog": catalog,
            "schema": schema,
            "model": model,
            "run_id": run_id,
            "use_case": use_case,
            "train_table": train_table,
            "freq": freq,
            "prediction_length": str(prediction_length),
            "backtest_length": str(backtest_length),
            "stride": str(stride),
            "metric": metric,
            "group_id": group_id,
            "date_col": date_col,
            "target": target,
            "num_nodes": str(num_nodes),
        },
    )
    print(f"[{i}/{len(active_models)}] Completed model: {model}")

print(f"\nAll {len(active_models)} models completed.")
```
