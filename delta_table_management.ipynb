```python
csv_file_path = "/Volumes/workspace/mmf_development/mmf_development_data/yahoo_stock.csv"
df = spark.read.csv(
    csv_file_path,
    header=True,           
    inferSchema=True,      
    sep=","            
)
df = df.toPandas()
df = df.rename(columns={"Adj Close": "Adj_Close"})
df['group_id'] = 1
df = spark.createDataFrame(data=df)
catalog = "workspace"
schema = "mmf_development"
table_name = f"{catalog}.{schema}.yahoo_stock"

df.write \
    .format("delta") \
    .mode("overwrite") \
    .option("mergeSchema", "true") \
    .saveAsTable(table_name) 
```
